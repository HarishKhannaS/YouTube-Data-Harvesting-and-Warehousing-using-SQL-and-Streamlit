```python
#Pipeline 1: Total Sales by Genre
pipeline = [
{"$unwind": "$sales"},
{"$group": {
"_id": "$genre",
"totalRevenue": {"$sum": {"$multiply": ["$price", "$sales.quantity"]}}
}}
]

results = records.aggregate(pipeline)
for result in results:
  print(result)
```

```python
#Pipeline 2: Average Monthly Sales for Each Book
pipeline = [
{"$unwind": "$sales"},
{"$group": {
"_id": "$title",
"averageMonthlySales": {"$avg": "$sales.quantity"}
}}
]

results = records.aggregate(pipeline)
for result in results:
  print(result)
```

```python
#Pipeline 3: Best Selling Author in February
pipeline = [
{"$unwind": "$sales"},
{"$match": {"sales.month": "February"}},
{"$group": {
"_id": "$author",
"totalFebruarySales": {"$sum": "$sales.quantity"}
}},
{"$sort": {"totalFebruarySales":-1}},
  {"$limit":3}
]
results = records.aggregate(pipeline)
for result in results:
  print(result)
```
